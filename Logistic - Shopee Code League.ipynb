```python
import copy
metroManila = {
    "metro manila": 3,
    "luzon": 5,
    "visayas": 7,
    "mindanao": 7,
}
luzon = copy.deepcopy(metroManila)
luzon['metro manila'] = 5
luzon['luzon'] = 5

visayas = copy.deepcopy(metroManila)
visayas['metro manila'] = 7
visayas['luzon'] = 7

mindanao = copy.deepcopy(visayas)

SLA = {
    "metro manila": metroManila,
    "luzon": luzon,
    "visayas": visayas,
    "mindanao": mindanao
}
```

```python
def checkEstimateSLA(seller,buyyer):
    seller = seller.lower()
    buyyer = buyyer.lower()
    
    sellerAdd = re.findall(r'metro manila|luzon|visayas|mindanao', seller)
    buyyerAdd = re.findall(r'metro manila|luzon|visayas|mindanao', buyyer)
    
    return SLA[sellerAdd[0]][buyyerAdd[0]]
```

```python
df['estimate'] = df.apply(lambda row: checkEstimateSLA(row.selleraddress[-15:],row.buyeraddress[-15:]), axis=1)
```

```python
df
```

```python
## Convert all date column to date
pick = pd.to_datetime(df['pick'], unit='s').dt.date
first_deliver = pd.to_datetime(df['1st_deliver_attempt'], unit='s').dt.date
second_deliver = pd.to_datetime(df['2nd_deliver_attempt'], unit='s').dt.date
```

```python
df['pick'] = pick
df['1st_deliver_attempt'] = first_deliver
df['2nd_deliver_attempt'] = second_deliver
```

```python
df
```

```python
public_holidays = ["2020-03-08", "2020-03-25", "2020-03-30", "2020-03-31"]
def get_busday_first(df):
    create_date = str(df['pick'])
    resolve_date = str(df['1st_deliver_attempt'])

    create_datetime = datetime.strptime(create_date, '%Y-%m-%d')
    resolve_datetime = datetime.strptime(resolve_date, '%Y-%m-%d')

    busday = np.busday_count(create_date, resolve_date, holidays=public_holidays, weekmask=[1,1,1,1,1,1,0])

    return busday
```

```python
df['1st_deliver_days'] = df.apply(get_busday_first, axis=1)
```

```python
def get_busday_second(df):
    create_date = str(df['1st_deliver_attempt'])
    resolve_date = str(df['2nd_deliver_attempt'])
    
    if resolve_date == 'NaT':
        resolve_date = create_date

    create_datetime = datetime.strptime(create_date, '%Y-%m-%d')
    resolve_datetime = datetime.strptime(resolve_date, '%Y-%m-%d')

    busday = np.busday_count(create_date, resolve_date, holidays=public_holidays, weekmask=[1,1,1,1,1,1,0])

    return busday 
```

```python
df['2nd_deliver_days'] = df.apply(get_busday_second, axis=1)
```

```python
df
```

```python
not_late_df = df[(df['1st_deliver_days'] <= df['estimate']) & (df['2nd_deliver_days'] <= 3)][['orderid']]
not_late_df['is_late'] = 0
not_late_df.head()
```

```python
late_df = df[~df['orderid'].isin(not_late_df['orderid'])][['orderid']]
late_df['is_late'] = 1
late_df.head()
```

```python
not_late_df.orderid.nunique() + late_df.orderid.nunique()
```

```python
res_df = pd.concat([late_df, not_late_df], axis=0)
res_df.head()
```

```python
res_df.to_csv("result.csv", index=False)
```
